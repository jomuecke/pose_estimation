```python
import os
import cv2

def resize_images_recursively(base_directory, target_width=1280, target_height=720):
    # Walk through the base directory recursively
    for root, dirs, files in os.walk(base_directory):
        for file_name in files:
            # Process only PNG images
            if file_name.endswith('.png'):
                file_path = os.path.join(root, file_name)
                
                # Load the image with OpenCV
                img = cv2.imread(file_path)
                
                if img is not None:
                    # Resize the image to the target dimensions (default: 1280x720)
                    img_resized = cv2.resize(img, (target_width, target_height))
                    
                    # Overwrite the original image with the resized image
                    cv2.imwrite(file_path, img_resized)
                    print(f"Resized and overwritten image: {file_path}")
                else:
                    print(f"Error loading image: {file_path}")

# Specify the base directory (update this with the path to your "extract_R17" folder)
base_directory = '/Users/jonasmucke/Desktop/R16'
resize_images_recursively(base_directory)
```
